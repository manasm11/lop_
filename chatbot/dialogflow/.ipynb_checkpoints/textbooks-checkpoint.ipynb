```python
import numpy as np
import pandas as pd
import xlsxwriter
from tqdm import tqdm
import yaml
```

```python
xls = pd.ExcelFile('Timetable_13_Mar_2021-67-97-converted.xlsx')
df = pd.read_excel(xls)
```

```python
for i in range(2, 8):
    df = pd.concat([pd.read_excel(xls, f'Table {i}'), df], ignore_index=True)
df
```

```python
df.to_excel('temp.xlsx')
```

```python
for c in df.columns: print(c)
```

```python
df.fillna('', inplace=True)


for row in df.index:
    cno = df.loc[row, 'COURSE NO']
    cnn = df.loc[row, 'Unnamed: 2'] + df.loc[row, 'Unnamed: 3']
    final_cno = ''
    if cno and cnn:
        final_cno = cno + ' ' + cnn
    df.loc[row, 'code'] = final_cno
    df.loc[row, 'textbook'] = df.loc[row, 'TEXT BOOK(S)'] + df.loc[row, 'Unnamed: 5']
for row in df.index:
    if not df.loc[row, 'code']:
        df.loc[row-1, 'textbook'] = df.loc[row-1, 'textbook'] + ' + '+df.loc[row, 'textbook']
#         if not df.loc[row-1, 'code']:
#             df.loc[row-2, 'textbook'] = df.loc[row-2, 'textbook'] + ' + '+df.loc[row-1, 'textbook']
    
del df['COURSE NO']
del  df['COM COD']
del df['Unnamed: 2']
del df['Unnamed: 3']
del df['MODE OF SUPPLY']
del df['Unnamed: 7']
del df['Unnamed: 9']
del df['Unnamed: 5']
del df['TEXT BOOK(S)']
df
```

```python
df.to_excel('temp3.xlsx', index=False)
```
